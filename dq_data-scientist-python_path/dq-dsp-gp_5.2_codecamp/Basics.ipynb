# Identifying Optimal Advertising Markets for an E-learning Company

## Introduction

As members of an e-learning company offering courses in programming, including web and mobile development and data science, we are interested in investing in advertisement to promote our product. The goal of this analysis is to determine the two best markets in which to advertise our e-learning courses in order to maximize our investment.

## Exploring Existing Survey Data

A useful method to identify optimal advertising choices is to conduct surveys for a number of different markets. This is, however, a costly and time consuming method. Instead, we can utilize ready-made survey data conducted for a similar company and relevant to our purpose.

For this analysis, we use a survey conducted in 2017 by the free e-learning platform [freeCodeCamp](https://www.freecodecamp.org), which offers courses primarily on web development and runs a popular [tech news publication](https://www.freecodecamp.org/news/). The [New Coder Survey](https://www.freecodecamp.org/news/we-asked-20-000-people-who-they-are-and-how-theyre-learning-to-code-fff5d668969/) was taken by tens of thousands of new coders with diverse interests covering a wide range of programming topics. The full survey data is available publicly on [GitHub](https://github.com/freeCodeCamp/2017-new-coder-survey).

Below, we briefly expore the dataset to familiarize ourselves with its structure and contents.

```python
# read in survey
import pandas as pd
survey = pd.read_csv('2017-fCC-New-Coders-Survey-Data.csv',
                     low_memory=False)
pd.set_option('display.max_columns', None)
```

```python
# get number of rows and columns
survey.shape
```

```python
# print first five rows
survey.head()
```

The survey dataset contains 18175 rows, each corresponding to a different survey respondent, and 136 columns, each representing a different survey question. The columns contain information related to the age, location, experience, interests, and career goals of each respondent. 

### Determining the Representativeness of our Sample

For the analysis at hand, i.e. identifying the best markets for advertising our programming e-learning courses, the population of interest is the set of all new coders who wish to learn the  subjects of the courses we offer. Specifically, we want to know the locations of the new coders, the locations with the most new coders, and the amount of money new coders are willing to spend on learning.

Before moving forward with our analysis using the existing freeCodeCamp New Coder Survey data, we need to make sure the sample we have is representative of our population of interest and that it contains the appropriate information to answer the questions above.

To assess the representativeness of our sample, we can look at the job role interests of the survey respondents. Since our e-learning courses focus primarily on web and mobile development, we want to ensure a large portion of the survey respondents are interested in these subjects. However, we also offer courses about a variety of other programming topics, so we also want the survey respondents to have a diverse set of job role interests overall. If both statements are true for the survey sample, we can conclude it is indeed representative of our population of interest.

```python
# unique job role interests
job_unique = survey['JobRoleInterest'].unique().size
job_unique
```

```python
# frequency distribution table (percentages) for job role interests
job_freq = survey['JobRoleInterest'].value_counts(normalize=True) * 100
job_freq
```

There are over 3000 unique values in the `JobRoleInterest` column, but many of these unique values consist of combinations of multiple individual subjects. This indicates that most people are interested in a wide variety of subjects.

Below, we zero in on the ten `JobRoleInterest` survey answers with the highest percentages. We see that both web and mobile developer are among the top ten job interests, along with data scientist, game developer, information security, and product manager.

```python
job_freq[0:10]
```

We can visualize the total percentage of survey respondents interested in each of the most popular subjects in the bar graph below.

```python
# get total counts for each individual popular job role interest
jobs = survey['JobRoleInterest'].dropna()  # drop null entries
job_roles = ['Web Developer', 'Mobile Developer', 'Data Scientist',
             'Game Developer', 'Information Security', 'Product Manager']
job_pcts = []
for job in job_roles:
    # percent of user with any job interests (i.e. who answered question)
    job_pcts.append(jobs.str.contains(job).sum() / jobs.size * 100)
```

```python
# plot percentages of popular job role interests
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1.25, 7.25, step=1.0)
plt.bar(x, job_pcts)
plt.xticks(x, job_roles, rotation = 45)
plt.xlim(1, 7.25)
plt.ylim(0, 100)
plt.xlabel('Job Role Interest')
plt.ylabel('Percentage')
plt.title('Percentages of Users Interested in Job Roles')
```

The bar graph above shows that over 80% of users with any job role interests are interested in web development, while nearly 40% are interested in mobile development. These two subjects are the most prevalent courses offered by our e-learning company. Meanwhile, over 20% of users are interested in data science and game development, and 10-20% of users are also interested in information security and product management. This shows that the sample dataset we're working with contains a representative set of coders with various interests relevant to our product. Therefore, we can consider it a representative sample of our population.

Now we are ready to answer the questions originally posed above:
- Where are the new coders located?
- What locations house the most new coders?
- How much money are new coders willing to spend on e-learning?

## New Coder Locations

For advertisement targeting purposes, we are interested in where a coder currently resides, as we can consider individual countries as individual markets. With this definition, our analysis goal becomes determining the two best countries in which to advertise. 

The number of survey participants living in a given country is a good indicator of the strength of that market, since the more potential customers in a market, the more chance for a larger number of purchases. Therefore, we will want to identify the two countries where the largest number of new coders, or survey participants, currently live.

To do so, we will investigate the frequency distribution table of the `CountryLive` column. This column contains the country in which each participant currently lives, regardless of their original country of origin. Since we based the representativeness of our sample off of the `JobRoleInterest` column information, we will drop all rows in our dataset where the entries in this column are null, i.e. where the participant did not indicate any specific job role interest. This will ensure our sample is representative of the population of users interested in our product.

```python
# drop rows without job interests
survey_rep = survey[survey['JobRoleInterest'].notnull()].copy()
```

```python
# generate frequency table for `CountryLive`
country_freq_abs = survey_rep['CountryLive'].value_counts()
country_freq_rel = survey_rep['CountryLive'].value_counts(normalize=True)
```

```python
country_freq_abs[0:10]
```

```python
country_freq_rel[0:10]
```

The two markets with the largest number of new coders are the United States and India. The United States is the largest market by far, accounting for nearly 50% of the entire sample. India, on the other hand, represents only about 8% of the users in the sample and is not much larger than the next few markets below it. While this information tells us we certainly want to advertise in the United States, it is worthwhile to investigate the data further before deciding on our second advertising market.

## Spending Money for Learning

Another metric to investigate when determining the strength of a market is the available money to be spent. A country with a large number of potential customers is not particularly profitable to us if those customers are unwilling to spend much money on programming courses. Since our company subscriptions sell by the month, we would like to know the amount of money for learning each coder is willing to spend per month. 

We will isolate this particular analysis to the following four countries: the United States, India, the United Kingdom, and Canada. When determining the largest potential spenders for our product, we restrict ourselves to these countries because they are the four largest markets. Additionally, they are all English-speaking countries, and our courses are written in English, so users from these countries are more likely to use our product.

The survey data contains a column labeled `MoneyForLearning` that describes the amount of money in America dollars spenty by the coder over all time, from the beginning of their coding experience until the time of the survey. Below, we use this information to calculate the average money spent per month by users in a given country in order to determine the markets with the largest potential spending power.

```python
# calculate money spent per month
survey_rep['MonthsProgramming'].replace(0,1, inplace=True)
survey_rep['MoneyPerMonth'] = survey_rep['MoneyForLearning'] / survey_rep['MonthsProgramming']
survey_rep['MoneyPerMonth'].isnull().sum()  # null values
```

Since we are interested spending money per month by country, we will only look at data where we have information in both of the `MoneyPerMonth` and `CountryLive` columns. In other words, we will drop all rows containing null entries in either of these columns to maintain the representativeness of our sample to our larger population.

```python
# remove rows with null values
survey_rep = survey_rep[survey_rep['MoneyPerMonth'].notnull()]
survey_rep = survey_rep[survey_rep['CountryLive'].notnull()]
```

```python
# group data by country and find mean money spent per user per month
country_mean = survey_rep.groupby('CountryLive').mean()
country_mean['MoneyPerMonth'][['United States of America',
                               'India',
                               'United Kingdom',
                               'Canada']]
```

Above, we find that users in the United States spend the most money on learning, with an average spending of over \$200 per month. Indian users are the next highest average spenders among the four largest English markets, spending over half of what Americans spend. Canada and the United Kingdom then come in third and fourth in terms of spending, with an average spending of approximately \$100 and \$50 a month, respectively. 

These results are somewhat surprising, since we would naively expect users from the United Kingdom and Canada to spend more than those in India based on the socio-economics of each country. It's possible there is not enough representative data for each country or that outliers exists which artificially shift the means. It's also possible that these results are correct and that India is in fact the second best market for our advertisements. Before drawing any conclusions, we analyze the data further.

## Looking for Outliers

Below, we generate box plots for the average money spent per user per month for the four countries of interest in order to visualize how the distributions compare and to identify any extreme outliers that may exist.

```python
# isolate four countries of interest
survey_rep_4 = survey_rep[survey_rep['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada')]

# box plots
import seaborn as sns
sns.boxplot(x = 'CountryLive', 
            y = 'MoneyPerMonth',
            data = survey_rep_4)
plt.title('Average Money Spent for Learning Per User Per Month', 
          fontsize=14)
plt.xlabel('Country of Residence')
plt.ylabel('Money spent per month (US dollars)')
plt.show()
plt.show()
```

Initially, we see two very extreme outliers for the United States, where users supposedly spend over $50,000 a month on learning. This seems unreasonable, so let's remove these outliers and then replot.

```python
# remove entries where `MoneyPerMonth` > 20000
survey_rep_4 = survey_rep_4[survey_rep_4['MoneyPerMonth'] < 20000]
```

```python
sns.boxplot(x = 'CountryLive', 
            y = 'MoneyPerMonth',
            data = survey_rep_4)
plt.title('Average Money Spent for Learning Per User Per Month', 
          fontsize=14)
plt.xlabel('Country of Residence')
plt.ylabel('Money spent per month (US dollars)')
plt.show()
```

Now we are able to more easily interpret the distributions in box plots. We still see a number extreme outliers in three of the four countries: several users in the United States reported spending an average of over \$6000 per month, a few users in India are outliers with spendings of over \$2500 a month, and one user in Canada reported an average spending of \$5000 per month. 

It's possible some or all of these data points are accurate, since students with large spendings could have attended bootcamps, which can be quite expensive, or could be avid learners who subscribe to many different programming media. Let's investigate these data points more closely to determine if their large values can be justified.

```python
# look at US outliers
us_outliers = survey_rep_4[
    (survey_rep_4['CountryLive'] == 'United States of America') & 
    (survey_rep_4['MoneyPerMonth'] >= 6000)]
us_outliers
```

```python
# look at India outliers
india_outliers = survey_rep_4[
    (survey_rep_4['CountryLive'] == 'India') &
    (survey_rep_4['MoneyPerMonth'] >= 2500)]
india_outliers
```

```python
# look at Canada outliers
canada_outliers = survey_rep_4[
    (survey_rep_4['CountryLive'] == 'Canada') &
    (survey_rep_4['MoneyPerMonth'] >= 5000)]
canada_outliers
```

Several of the spending outliers in the dataset have attended bootcamps, which justifies their large spending amounts. However, many of the outliers have not attended bootcamps and do not appear to have any other justifications for their large spendings. Therefore, we will drop any outliers that have not attended bootcamps. We will also drop any outliers where the user has only been programming for one or two months, since we do not have enough monthly data to determine if their initial spending is indicative of their overall monthly spending habits.

```python
# remove outliers who have not attended bootcamp
rm_us_out = ((survey_rep_4['CountryLive'] == 'United States of America') & 
             ((survey_rep_4['MoneyPerMonth'] < 6000) | 
              (survey_rep_4['AttendedBootcamp'] == 1.0)))
rm_ind_out = ((survey_rep_4['CountryLive'] == 'India') & 
              ((survey_rep_4['MoneyPerMonth'] < 2500) | 
              (survey_rep_4['AttendedBootcamp'] == 1.0)))
rm_can_out = ((survey_rep_4['CountryLive'] == 'Canada') & 
             ((survey_rep_4['MoneyPerMonth'] < 5000) | 
              (survey_rep_4['AttendedBootcamp'] == 1.0)))
survey_rep_4 = survey_rep_4[rm_us_out | rm_ind_out | rm_can_out | 
                            (survey_rep_4['CountryLive'] == 'United Kingdom')]
```

```python
# remove outliers with less than 3 months programming experience
rm_us_new = ((survey_rep_4['CountryLive'] == 'United States of America') & 
             ((survey_rep_4['MoneyPerMonth'] < 6000) | 
              (survey_rep_4['MonthsProgramming'] >= 3.0)))
rm_ind_new = ((survey_rep_4['CountryLive'] == 'India') & 
              ((survey_rep_4['MoneyPerMonth'] < 2500) | 
               (survey_rep_4['MonthsProgramming'] >= 3.0)))
rm_can_new = ((survey_rep_4['CountryLive'] == 'Canada') & 
              ((survey_rep_4['MoneyPerMonth'] < 5000) | 
               (survey_rep_4['MonthsProgramming'] >= 3.0)))
survey_rep_4 = survey_rep_4[rm_us_new | rm_ind_new | rm_can_new | 
                            (survey_rep_4['CountryLive'] == 'United Kingdom')]
```

```python
sns.boxplot(x = 'CountryLive', 
            y = 'MoneyPerMonth',
            data = survey_rep_4)
plt.title('Average Money Spent for Learning Per User Per Month', 
          fontsize=14)
plt.xlabel('Country of Residence')
plt.ylabel('Money spent per month (US dollars)')
plt.show()
```

Now we can recompute the mean spending per month per user by country of residence.

```python
# group data by country and find mean money spent per user per month
country_mean_4 = survey_rep_4.groupby('CountryLive').mean()
country_mean_4['MoneyPerMonth'][['United States of America',
                                 'India',
                                 'United Kingdom',
                                 'Canada']]
```

## Choosing the Best Advertising Markets

After removing outliers, we found the average monthly spending of users in each of the four largest English-speaking markets. The largest spenders are in the United States, with approximately \$150 a month. The next largest spenders are Canadians, with nearly \$100 spent a month. Indian users are the third largest spenders, with about \$65 monthly spendings.

Based on spending power alone, we might be tempted to advertise in Canada over India, but we also to include market size when considering where to advertise. Above, we calculated the size of each market in terms of the number of users per country. We found the largest market by far to be the United States, followed by India and then Canada. 

We clearly want to advertise in the United States, since it both the largest and highest-spending market. Choosing a second market is a more nuanced decision. India's market is over twice the size of Canada's, but Canadians spend nearly \$30 more per month. It is unclear which metric, size or money, is a better indicator of the strength of the market, so we can not immediately conclude whether India or Canada is a better market for our advertisements. If we have the money to advertise in three countries, we would recommend choosing the US, India, and Canada. If we must choose only two without any further information, we would recommend the US and India. While India spends less than Canada on average, the market there is larger, and the average spending is still above our monthly subscription price.

Without further information, we would likely send our findings along to the marketing team, whose unique domain knowledge could help them to make the best final advertising decision.

## Conclusion

We can solidly conclude that we should advertise our e-learning courses in the United States, which is the largest and highest spending English-speaking market. Beyond that, either India or Canada would make the second best market in which to advertise, but we leave the final decision up to the marketing team.
